```python
genre = dados.groupby(['genre']).size()
genre
```

```python
#amostra = dados.groupby('genre', group_keys=False).apply(nlargest(250, ['worldwide_gross']))

#amostra

genero = dados.groupby('genre').apply(pd.DataFrame.sample, frac=1)

action = genero[genero.genre=='Action']
adventure = genero[genero.genre=='Adventure']
comedy = genero[genero.genre=='Comedy']
drama= genero[genero.genre=='Drama']
horror = genero[genero.genre=='Horror']


amostra = action.nlargest(250, ['worldwide_gross']) 
amostra = pd.concat([amostra, adventure.nlargest(250, ['worldwide_gross'])], ignore_index=True)
amostra = pd.concat([amostra, comedy.nlargest(250, ['worldwide_gross'])], ignore_index=True)
amostra = pd.concat([amostra, drama.nlargest(250, ['worldwide_gross'])], ignore_index=True)
amostra = pd.concat([amostra, horror.nlargest(250, ['worldwide_gross'])], ignore_index=True)
amostra
```

```python
amostra['Perc'] = (amostra['worldwide_gross'] / amostra['production_budget']) *100

amostra
```

```python
amostra['Tamanho'] = amostra['Perc'] / 100

amostra
```

```python
genero = amostra.groupby('genre').apply(pd.DataFrame.sample, frac=1)

horror = genero[genero.genre=='Horror']

janela = plt.figure(figsize=(4,4))
histograma= janela.add_axes([0,0,1,1])
histograma.hist(horror['Tamanho'])
```
